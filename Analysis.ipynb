```python
df.groupby('Customer Segment')[['Revenue (ZAR)', 'Gross Profit (ZAR)']].mean()
```

```python
total_revenue = df['Revenue (ZAR)'].sum()
total_profit = df['Gross Profit (ZAR)'].sum()
profit_margin = (total_profit / total_revenue) * 100
print(f"Overall Profit Margin: {profit_margin:.2f}%")
```

```python
# Top 500 products by total revenue
df.groupby('Product Name')['Revenue (ZAR)'].sum().nlargest(500)

# Top 500 products by gross profit margin
prod_perf = df.groupby('Product Name')[['Revenue (ZAR)', 'Gross Profit (ZAR)']].sum()
prod_perf['Margin %'] = (prod_perf['Gross Profit (ZAR)'] / prod_perf['Revenue (ZAR)']) * 100
prod_perf.sort_values(by='Margin %', ascending=False).head(500)
```

```python
# Group by Month to see sales trajectory
df.set_index('Date').resample('M')['Revenue (ZAR)'].sum()
```

```python
# Creates a heatmap-ready grid of Revenue
pd.crosstab(df['Customer Segment'], df['Category'], values=df['Revenue (ZAR)'], aggfunc='sum')
```

```python
# Average quantity per order across different segments
df.groupby('Customer Segment')['Quantity Sold'].mean()

# Check correlation between Quantity Sold and Gross Profit
df['Quantity Sold'].corr(df['Gross Profit (ZAR)'])
```

```python
# Calculate COGS as a % of Revenue by Category
cat_costs = df.groupby('Category')[['Revenue (ZAR)', 'COGS (ZAR)']].sum()
cat_costs['COGS % of Revenue'] = (cat_costs['COGS (ZAR)'] / cat_costs['Revenue (ZAR)']) * 100
cat_costs.sort_values(by='COGS % of Revenue', ascending=False)
```
